```python
# basic
import numpy as np
import pandas as pd

# matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

# seaborn
import seaborn as sns
```

```python
ironman_np_x = np.linspace(1, 14, 100) #產生一個1~14的等差數列
ironman_np_y1 = np.sin(ironman_np_x)
ironman_np_y2 = np.sin(ironman_np_x+2)*1.25
```

```python
def drawPlot():
    plt.plot(ironman_np_x, ironman_np_y1)
    plt.plot(ironman_np_x, ironman_np_y2)
```

```python
fig = plt.figure()
drawPlot()
fig.savefig('day19_01.png')
```

```python
fig = plt.figure()
sns.set_style('whitegrid')
drawPlot()
fig.savefig('day19_02.png')
```

```python
def drawPlotFour():
    for i in range(4):
        plt.plot(ironman_np_x, np.sin(ironman_np_x+i)*(i+0.75), label="sin(x+i)*(i+0.75)")
    plt.legend()
```

```python
fig = plt.figure(figsize=(10, 6))
drawPlotFour()
fig.savefig('day19_04.png')
```

```python
with sns.axes_style('white'):
    fig = plt.figure(figsize=(10, 6))
    drawPlotFour()
    fig.savefig('day19_05.png')
```
