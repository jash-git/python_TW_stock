```python
# basic
import numpy as np
import pandas as pd

# matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

# seaborn
import seaborn as sns
```

```python
ironman_series = pd.Series(np.random.randn(1000))
fig = plt.figure()
ironman_series.plot(kind='hist')
fig.savefig('day18_01.png')
```

```python
fig = plt.figure()
ironman_series.plot(kind='kde')
fig.savefig('day18_02.png')
```

```python
fig = plt.figure()
sns.distplot(ironman_series)
fig.savefig('day18_03.png')
```

```python
sns.kdeplot(ironman_series)
```

```python
ironman_df = sns.load_dataset('flights')
ironman_df = ironman_df.pivot(index='month', columns='year', values='passengers')
```

```python
fig = plt.figure()

ironman_df.sum().plot(kind="bar")

fig.savefig('day18_04.png')
```

```python
ironman_df.sum().plot(kind="bar")
```

```python
fig = plt.figure()
sns.barplot(ironman_df.sum().index,ironman_df.sum().values)
fig.savefig('day18_05.png')
```

```python
fig = plt.figure()
sns.heatmap(ironman_df, annot=True , fmt='d')
fig.savefig('day18_06.png')
```
